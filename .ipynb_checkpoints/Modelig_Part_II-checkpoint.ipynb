## Fit a logistic regression model and store the class predictions.

```python
#Build and train a logistic regression model.
###Select 2 features for your feature matrix, $X$
##use train_test_split() with random_state=1 and the default split.
#$y$ will remain the same glass_df['household']
#Evaluate the model with .score().
```

```python

# We already imported this above. It's here just for reference

# from sklearn.linear_model import LogisticRegression


feature_cols = ['Income','GenHlth','MentHlth','PhysHlth','DiffWalk']
X = data['all'][feature_cols]
y =data['all']["Diabetes_binary"] 

#step 1: instantiate
logreg = LogisticRegression()


#step 2 fit
logreg.fit(X,y)

#step 3: use features to predict & assign to variable
type_pred = logreg.predict(X)
```

```python

type_pred[:100]
```


#### Plot the predicted class using the logistic regression  for menhealth predictions  

```python
df_sort_desc = data['all'][['MentHlth','Diabetes_binary']].sort_values('MentHlth',ascending=True)
```

```python
df_sort_desc.head
```

```python
 
# Plot the class predictions.
plt.scatter(df_sort_desc['MentHlth'], df_sort_desc['Diabetes_binary'])
plt.plot(df_sort_desc['MentHlth'], type_pred, color='red')
plt.xlabel('MentHlth')
plt.ylabel('Diabetes_binary');
```

```python
logreg.predict_proba(X) [3]
```

predicted probabilities instead of just the class predictions, to understand how confident we are in a given prediction?

Using the built-in .predict_proba() function, examine the predicted probabilities for the first handful of rows of X.

```python

# Store the predicted probabilities of class 1.

data['all']['diagnose_pred_prob'] = logreg.predict_proba(X)[:, 1]
```

```python
data['all'].tail()
```

```python
 
# Plot the predicted probabilities.
plt.scatter(data['all'] ['MentHlth'], y) 
plt.plot(data['all']['MentHlth'] ,data['all']['diagnose_pred_prob'], color='red')
plt.xlabel(data['all']['Diabetes_binary'])
plt.ylabel(data['all']['Diabetes_binary']);

#this is a non linear relationship  Household is not increasing in a straighline
```

```python

# Examine some example predictions.
#print predictions for 1-3. use a for loop.
print(f'Predicted Probabilities for each class:')
for i in range(1,4):
    print(logreg.predict_proba(np.array(i).reshape(-1,1)))
```
